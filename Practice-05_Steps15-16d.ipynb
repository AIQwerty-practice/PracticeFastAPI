### Step 1 to 5 ☑️

- Clone repository <br>
- git clone https://github.com/inskillflow/demo_api_1_simple_fastapi_app.git <br>

### Step 2 — Create the virtual environment ☑️
- python -m venv venv <br>
- .venv\Scripts\activate <br>
- Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser <br>

### Step 3 - Install dependencies ☑️<br>

- Upgrade pip first ☑️<br><br>
python -m pip install --upgrade pip<br><br>

- Install all packages ☑️<br><br>
- pip install -r requirements.txt<br>
- Verify : pip list<br>

### Step 4 — Run the FastAPI server <br>
- uvicorn main:app --reload <br>

### Step 5 — Run the Streamlit frontend <br><br>
- venv\Scripts\activate <br>
- streamlit run frontend.py <br>

- The server starts at http://127.0.0.1:8000 <br>

URL	What you see <br>
http://127.0.0.1:8000/	Root — confirms the API is running <br>
http://127.0.0.1:8000/tasks	All tasks (JSON) <br>
http://127.0.0.1:8000/docs	Swagger UI — interactive documentation <br>
http://127.0.0.1:8000/redoc	ReDoc — clean read-only docs <br>


### Step 6 — Test the API <br> 

- Option A — Swagger UI (no setup needed)<br> <br> 
- Go to http://127.0.0.1:8000/docs in your browser. <br>  
- Click any endpoint → Try it out → fill in values → Execute.
